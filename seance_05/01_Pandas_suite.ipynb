## Données Manquantes :


### .isna() / .notna()

Pandas a des méthodes spéciales pour traiter les données manquantes. Comme vous l'avez peut-être déjà remarqué, les lignes vides dans un fichier CSV apparaissent comme `NaN` dans un DataFrame Pandas.

Pour filtrer et compter le nombre de valeurs manquantes/non manquantes dans un jeu de données, nous pouvons utiliser les méthodes spéciales `.isna()` et `.notna()` sur un objet DataFrame ou Series.



```python
slaves_df.iloc[:, 3:].head(2)
```

```python
slaves_df['percent_women'].notna()
```

Par exemple, le pourcentage de femme dans chaque navire :

```python
slaves_df[slaves_df['percent_women'].notna()]
```

Pour compter explicitement le nombre de lignes vides, nous pouvons utiliser la méthode `.value_counts()`.


```python
slaves_df['percent_women'].isna().value_counts()
```

```python
slaves_df['percent_women'].isna().value_counts(normalize=True)
```

### .count()

Comme la méthode `.count()` exclut toujours les valeurs NaN, nous pouvons également compter le nombre de valeurs dans chaque colonne, puis diviser par le nombre total de lignes dans chaque colonne (`len()`) pour trouver le pourcentage de données non vides dans chaque colonne.

```python
slaves_df.count() / len(slaves_df)
```

Si nous le souhaitions, nous pourrions remplir les valeurs NaN dans le DataFrame avec une valeur différente en utilisant la méthode `.fillna()`.

## Groupby Colonnes

La fonction `.groupby()` de Pandas nous permet de regrouper des données et d'effectuer des calculs sur ces groupes.

Par exemple, Jennifer Morgan explique comment certaines nations ont enregistré plus d'informations sur le genre des personnes réduites en esclavage à bord de leurs voyages que d'autres nations. Pour voir la répartition des informations sur le genre par nation, nous pouvons utiliser la fonction `.groupby()`.

La première étape pour utiliser `groupby` est de taper le nom du DataFrame suivi de `.groupby()` avec la colonne que nous voulons agréger, telle que "national_affiliation".


```python
slaves_df.groupby('flag')
```

Cette action va créer un objet GroupBy. Nous pouvons effectuer des calculs sur ces données groupées, comme compter le nombre de valeurs non vides dans chaque colonne pour chaque nation.

```python
slaves_df.groupby('flag').count()
```

On peut isoler la colonne "percent_women"

```python
slaves_df.groupby('flag').count()['percent_women']
```

```python
slaves_df.groupby('flag')['percent_women'].count().sort_values(ascending=False)
```

```python
slaves_df.groupby('flag')['percent_women'].count()\
.sort_values(ascending=False).plot(kind='bar', title='Trans-Atlantic Slave Trade (Americas): \n Slave Voyages with Recorded Gender Information')
```

### Graphique sur le temps: 

```python
slaves_df.groupby('year_of_arrival')['total_women'].sum()
```

```python
total_women_by_year = slaves_df.groupby('year_of_arrival')['total_women'].sum()
```

```python
total_women_by_year.plot()
```

```python
total_men_by_year = slaves_df.groupby('year_of_arrival')['total_men'].sum()
```

```python
ax_w = total_women_by_year.plot(kind='line', legend= True,
                              title="Trans-Atlantic Slave Trade (Americas):\nTotal Number of Enslaved Women on Voyages")
total_men_by_year.plot(ax=ax_w, legend=True)
```

Enfin, nous pouvons également ajouter le nombre total de personnes réduites en esclavage embarquées lors des voyages, offrant ainsi une perspective sur la quantité d'informations sur le genre que nous avons par rapport au nombre total de voyages.

```python
total_embarked_by_year = slaves_df.groupby('year_of_arrival')['total_embarked'].sum()
```

```python
ax = total_women_by_year.plot(kind='line', label="Total Women", legend= True, title="Trans-Atlantic Slave Trade (Americas):\nTotal Number of Enslaved Women on Voyages")
total_men_by_year.plot(ax=ax, label="Total Men", legend=True)
total_embarked_by_year.plot(ax=ax, label="Total Embarked", legend=True)
```
