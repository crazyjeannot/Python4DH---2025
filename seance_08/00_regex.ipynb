## Exercice 1 – Répliques de dialogue

Dans le texte, les dialogues commencent par un tiret `–`.

1. Écris une expression régulière qui sélectionne **toutes les lignes de dialogue**, c’est-à-dire toutes les lignes qui commencent par `–`.
2. Utilise `re.findall` pour obtenir la liste de toutes ces lignes à partir de la variable `texte`.
3. Affiche le nombre total de répliques trouvées.

*(Indice : pense à utiliser le flag `re.MULTILINE`.)*

```python
import re
```

```python
repliques = re.findall(r"^–.*$", txt, flags=re.MULTILINE)

print(len(repliques))
# On affiche les 5 premières répliques trouvées
for r in repliques[:5]:
    print(r)
```

## Exercice 2 – Phrases exclamatives

On veut repérer les **phrases qui se terminent par un point d’exclamation** `!`.

1. Propose une définition naïve d’une phrase :

   > une suite de caractères qui commence juste après un `.` `?` ou `!` (ou le début du texte)
   > et qui se termine par un `!`.
2. Écris une expression régulière qui sélectionne ces phrases exclamatives.
3. Utilise `re.findall` pour extraire toutes les phrases exclamatives du texte et les afficher.


```python
pattern_exclamatives = r"[^\.?!]*!"
phrases_exclam = re.findall(pattern_exclamatives, txt)

print("\nExercice 2 – Nombre de phrases exclamatives :", len(phrases_exclam))
print("Exemples de phrases exclamatives :")
for p in phrases_exclam[:5]:
    print("  ", p.strip())

```


## Exercice 3 – Noms propres / Noms de lieux

On veut repérer les **mots qui commencent par une majuscule** (noms propres, lieux, personnages…).

1. Écris une expression régulière qui sélectionne les mots :

   * qui commencent par une lettre majuscule (`A`–`Z`),
   * suivie de plusieurs lettres minuscules (éventuellement avec des accents),
   * éventuellement avec un tiret au milieu (par ex. `Saint-Nicolas`).
2. Utilise `re.findall` pour obtenir une liste de tous les candidats.
3. Affiche les 20 premiers résultats et repère des lieux connus dans le texte.


```python

pattern_noms = r"[A-Z][a-z]*[-].[a-z]*|[[A-Z][a-z]*]"
pattern_noms = r"\b[A-Z][a-z]{2,}\b"
noms_propres = re.findall(pattern_noms, txt)

print("\nExercice 3 – Nombre de noms propres (candidats) :", len(noms_propres))
print("Les 20 premiers :")
print(noms_propres[:20])
```


## Exercice 4 – Citations entre guillemets français « … »

Le texte contient des citations entre guillemets français `«` et `»`.

1. Écris une expression régulière qui **capture tout ce qui se trouve entre** `«` et `»`.
2. Utilise `re.findall` pour récupérer toutes ces citations dans une liste.
3. Affiche chaque citation sur une nouvelle ligne.


```python
pattern_citations = r"«(.*?)»"
citations = re.findall(pattern_citations, txt, flags=re.DOTALL)

print("\nExercice 4 – Nombre de citations trouvées :", len(citations))
print("Exemples de citations :")
for c in citations:
    print("  «" + c.strip() + "»")

```
