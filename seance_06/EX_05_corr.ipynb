
## inverser une chaîne mot par mot

Écrire une fonction `inverser_phrase(phrase)` qui inverse les lettres de chaque mot sans changer l’ordre des mots.

```python
phrase = "Les humanités numériques sont fascinantes"
# → "seL sétinamuh seuqirémun tnos setnanicisaf"
```

````python
phrase = "Les humanités numériques sont fascinantes"

def inverser_phrase(phrase):
    p= ''
    split_phrase= phrase.split()
    l = len(split_phrase)
    i=0
    for mot in split_phrase : 
        if i == l-1 : 
            p+=inverser_mot(mot)
        else : 
            p+= inverser_mot(mot) + " "
            i+=1
    return p 

def inverser_mot(mot): 
    rev_mot = list(mot)
    rev_mot.reverse()
    s= ''
    for lettre in rev_mot : 
        s+=lettre
    return s
    
````

````python
inverser_phrase(phrase)
````

````python
import numpy as np
def operation_matrices(n) : 
    A = np.random.randint(0,9,(n,n)),
    B = np.random.randint(0,9,(n,n))
    return A+B, A*B, np.dot(A,B)
````

### Quelle est la durée moyenne des missions par pays ?

````python
#NPO de convertir les dates en format date (vs str)

df_tsv["Date de départ"] = pd.to_datetime(df_tsv["Date de départ"], format="%d/%m/%Y",# errors="coerce")
)
df_tsv["Date de retour"] = pd.to_datetime(df_tsv["Date de retour"], format="%d/%m/%Y",# errors="coerce")
)
df_tsv["Durée"] = (df_tsv["Date de retour"] - df_tsv["Date de départ"]).dt.days

duree_moyenne = df_tsv.groupby("Pays de départ")["Durée"].mean().sort_values(ascending=False)
````

````python
duree_moyenne
````

````python
import matplotlib.pyplot as plt

duree_moyenne.plot(kind="bar", title="Durée moyenne des missions par pays", ylabel="Durée moyenne (jours)", figsize=(10, 5))
plt.tight_layout()
plt.show()

#duree_moyenne
````
